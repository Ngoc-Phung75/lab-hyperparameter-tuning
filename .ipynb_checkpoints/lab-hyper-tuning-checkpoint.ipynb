```python
df.drop(columns=['HomePlanet'], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

numerical_features = df.select_dtypes(include=['int64', 'float64'])
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(numerical_features), columns=numerical_features.columns)
le = LabelEncoder()
df['Destination'] = le.fit_transform(df['Destination'])
```

```python
from sklearn.impute import SimpleImputer

# Step 1: Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object', 'bool']).columns

# Step 2: Impute numerical columns
num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Step 3: Impute categorical columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# Step 4: Check for nulls
print(df.isnull().sum())
```

```python
#y = target
#X = features

y = df['FoodCourt']
X = df.drop(columns=['FoodCourt'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
# Loop through float columns and convert them to int if all values are whole numbers
for col in df.select_dtypes(include='float'):
    if (df[col] % 1 == 0).all():
        df[col] = df[col].astype(int)
```

```python
#Define the hyperparameter grid

paramgrid = {
    'n_estimators': [100, 200, 300],            # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],            # Maximum depth of tree
    'min_samples_split': [2, 5, 10],            # Min samples to split an internal node
    'min_samples_leaf': [1, 2, 4],              # Min samples at a leaf node
    'bootstrap': [True, False]                  # Whether bootstrap samples are used
}
```
